```python
#(1) From HW2_titanic.csv, read into the data in Pandas: 2pts

import pandas as pd
import matplotlib.pyplot as plt
titanic = pd.read_csv(r"C:\Users\anjal\Desktop\Validation\HW2_titanic.csv")
titanic
```

```python
#(2) Understand the meaning of each variable: 2pts

titanic.info(memory_usage='deep')
print("\ncolumn length:",len(titanic.columns))
print("\nrow length:",len(titanic.index))
print("The columns are:",titanic.columns)
print(titanic.shape)
```

```python
#(3) View all numeric class variables and their associated statistics: 2pts
titanic.describe()
```

```python
#(1) Number of survivors and deaths: 2pts

from pandas import DataFrame
import pandas as pd
titanic['Survival'] = titanic.Survived.map({0 : 'Died', 1 : 'Survived'}) 
df1=titanic['Survival'].value_counts()
print("\nDisplaying the number of survivors and deaths:\n",df1)
```

```python
#(2) Distribution of survivors and deaths at each level by a DataFrame: 4pts

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
titanic['Survival'] = titanic.Survived.map({0 : 'Died', 1 : 'Survived'}) 
df3 = titanic.groupby(['Survival', 'Pclass','Survived']).size().reset_index(name='count') 
result = pd.DataFrame(df3.pivot("Pclass", "Survival", "count"))
print("\n",result)
```

```python
#(1) Use a bar plot to show the distribution of survivors and deaths at each level: each level (Pclass) as x-axis, number of people as y-axis. In each same bar, "Survived" is on the top, and "Died" is on the bottom. 4pts

import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np 
titanic['Survival'] = titanic.Survived.map({0 : 'Died', 1 : 'Survived'}) 
df3 = titanic.groupby(['Survival', 'Pclass','Survived']).size().reset_index(name='count') 
plotdata = df3.pivot("Pclass", "Survival", "count")
graph = plotdata.plot.bar(stacked=True)
plt.xlabel('Pclass')
plt.ylabel('Number of people')
plt.title("Survivors and deaths distribution at each levels")

#(2) Show the legend as "Died" and "Survived". 1pt

plt.legend(loc='upper left')
plt.show()
```
